```python
# p.155
# 키워드 인수################################
def sum_range(begin, end, step=1) :
    sum = 0
    for n in range(begin, end, step) :
        sum += n
    return sum
##############################################
sum_range(end=10, begin=1, step=2)
```

```python
# p.183     Counter class
class Counter :  #클래스는 자기자신일수도, 부모일 수도 있다. 모든 클래스는 object조상님을 받음
  def reset(self) :
    self.count = 0
  def increment(self) :
    self.count += 1
  def get(self) :
    return self.count
  def __str__(self) :             # str메소드
    return "counter : " + str(self.count)
####################################################
```

```python
counter = Counter()  # 이 공간은 지금부터 카운터라고 지칭하자
counter.reset()
counter.increment()
print(counter.get())
print(counter)
```

```python
# Car 클래스 : speed, geer, color ========================================================
class Car :
  #생성자
  def __init__(self, speed, gear, color="black" ) :
    self.speed = speed
    self.gear = gear
    self.color = color
  #설정자(set) : setSpeed, setGear, setColor
  def setSpeed(self, speed) :
    self.speed = speed
  def setGear(self, gear) :
    self.gear = gear
  def setColor(self, color) :
    self.color = color
  #접근자(get) : getSpeed, getGear, getColor
  def getSpeed(self) :
    return self.speed
  def getGear(self) :
    return self.gear
  def getColor(self) :
    return self.color 
  # __str__
  def __str__(self) :
    return "speed : {}, gear : {}, color : {}".format(self.speed, self.gear, self.color)
#==========================================================================================
```

```python
car = Car(100, 3, "white")
print(car)
car.setSpeed(200)
print(car)
car.setGear(30)
print(car)

```

```python
# cat class =======================================
class Cat :
  #생성자
  def __init__(self, name, age) :
    self.name = name
    self.age = age

  #설정자(set) : setName, setAge
  def setName(self, name) :
    self.name = name
  def setAge(self, age) :
    self.age = age

  #접근자(get) : getName, getAge
  def getName(self) :
    return self.name
  def getAge(self) :
    return self.age

 # 메소드 추가
  def yaong(self) :
    print("{}가 야옹".format(self.name))

  # 연산자 중복
  def __eq__(self, othercat) :
    return self.age == othercat.age

  # __str__
  def __str__(self) :
    return "name : {}, age : {}".format(self.name, self.age)
#==========================================================================================
```

```python
cat = Cat("Missy", 3)
print(cat)
cat.setName("Lucky")
print(cat)
cat.setAge(5)
print(cat)
```
