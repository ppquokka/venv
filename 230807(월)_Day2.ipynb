```python
plt.rcParams['font.family'] = 'Malgun Gothic'
import csv
import matplotlib.pyplot as plt

f = open('census.csv', encoding='cp949')
data = csv.reader(f)
result = []   # 리스트생성
# 데이터가 행렬의 2차원 데이터이다
for row in data:           
  if '문래동' in row[0] :        # '문래동'이 포함된 행정구역 찾기     
    for i in row[3:] :         # 0세 부터 모든 연령 반복하기  
      result.append(int(i))      # 해당 연령의 인그수 리스트에 순서대로 저장하기
plt.plot(result)
plt.title('2023년 7월 인구 현황')
plt.xlabel('연령')                # x축 이름
plt.ylabel('인구수')              # y축 가로쓰기 이름
# h=plt.ylabel('인구수')          # y축 가로쓰기 이름
# h.set_rotation(0)               # 해당 항목을 가로로 표기
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('census.csv', encoding='cp949')
data = csv.reader(f)
result = []   # 리스트생성
dong = input("찾고싶은 동(또는 구):")
# 데이터가 행렬의 2차원 데이터이다
plt.rcParams['font.family'] = 'Malgun Gothic'
for row in data:           
  if dong in row[0] :        # '문래동'이 포함된 행정구역 찾기     
    for i in row[3:] :         # 0세 부터 모든 연령 반복하기  
      result.append(int(i))      # 해당 연령의 인그수 리스트에 순서대로 저장하기
plt.plot(result)
plt.title('2023년 7월 인구 현황')
plt.xlabel('연령')                # x축 이름
#plt.ylabel('인구수')              # y축 가로쓰기 이름
h=plt.ylabel('인구수')          # y축 가로쓰기 이름
h.set_rotation(0)               # 해당 항목을 가로로 표기
plt.show()
```
