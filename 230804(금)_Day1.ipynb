### 연산자

```python
# 산술연산자 : +, -, *, /, //(정수 몱), %(나머지), **(거듭제곱)
num1 =10
num2 = 20
result = num1 + num2 #더하기
result
result = num1 - num2 #빼기
result
result = num1 * num2 # 곱하기
result
result = num1 / num2 #나누기
result
result = num1 // num2 #나누기(정수, 몫)
result
result = num1 ** num2 #거듭제곱
result
result = num1 % num2 # 나머지
result

# 나머지 연산
if num1 % num2 == 0 :    # == 은 비교연산자, = 대입연산자(할당)
    print("짝수입니다.")
else : 
    print("홀수입니다.")
```

```python
# 관계연산자 : Bool(true / False)로 변환
num1 = 10
num2 = 20
result = num1 > num2
result
result = num1 < num2
result
result = num1 >= num2
result
result = num1 == num2
result
result = num1 != num2
result
```

```python
# 부울연산자, 논리연산자 : Bool(true / false)로 변환
# age > 30 and toeic > 700
toeic = int(input("토익점수 :"))
age = int(input("나이 :"))
# and(모두 만족할 때), or(어느 하나라도 만족시), not(반대로)
if age > 30 and toeic > 700 :
    print("합격입니다.")
else :
    print("불합격 입니다.")
```

## 조건문

```python
import random
lotto = []
count = 1

while True:
        if count > 7:
            break

        num = random.randrange(1, 45+1)
        if num not in lotto:
            lotto.append(num)
            count = count + 1

print(lotto)
```

```python
# 사용자에게 점수를 입력받아 학점을 체크해주세요.
score = int(input("점수 입력 :"))
if (score >= 90) :
    print('A')
elif (score >= 80) :
    print('B')
elif (score >= 70) :
    print('C')
elif (score >= 60) :
    print('D')
else :
    print('F')
```
