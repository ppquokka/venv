```python
# 파일 입력 (open, read, close)
inFp, inStr, inStr2 = None, [], []
inFp = open("data1.txt","r",encoding="utf-8")
inStr = inFp.readlines()
print(inStr)
#시퀀스인 inStr에서 각 문자열 원소(str)에서
for str in inStr :
  #rogod answk(\n)을 strip해서 inStr2에 append후
  inStr2.append(str.rstrip("\n").rstrip())
#inStr2를 출력해주세요
print(inStr2)
inFp.close()
```

```python
# 파일 쓰기 (open, write, close)
outFp = None
outStr = ""
outFp = open("data2.txt","w",encoding="utf-8")

while True :
  #1. input 받는다
  outStr = input("내용입력:")
  if outStr == "":
    break
  outFp.writelines(outStr + "\n")

outFp.close()
```

```python
import pymysql
# 1. DB에 연결한다
conn = pymysql.connect(host="localhost", db="sqldb", user="root", password="1234")

# 2.SQL을 위한 커서를 만든다
cur=conn.cursor()

#3. 커서를 execute한다.
sql = "select * from member"
cur.execute(sql)

#4.fetchall을 통해 전체 내용을 가져온다
result = cur.fetchall()
for line in result :
    id = line[0]
    pw = line[1]
    print(f"id : {id}, pw : {pw}")

#5.commit하고 close한다
conn.cnommit()
conn.close()
```

```python
# 정답!
#DB의 member table에 있는 내용을 읽어들여서,
# member.csv 파일에 저장해주세요
# aaa, 111 : join
'''
aaa, 111
bbb, 222
ccc, 333
'''

import pymysql
# DB 연결
outFp = open("member.csv", "w", encoding="utf-8")
conn = pymysql.connect(host="localhost", db="sqldb", user="root", password="1234")

# SQL 쿼리 실행
sql = "select * from member"
cur=conn.cursor()
cur.execute(sql)

# 결과를 가져와서 csv 파일에 저장
result = cur.fetchall()
with open("member.csv", "w") as file:
    for line in result :
        id = line[0]
        pw = line[1]
        outFp.writelines(f"{id},{pw}"+"\n")

# 연결 닫기
outFp.close()
cur.close()
conn.close()
```

```python
# line = cur.fetchone()  1씩 가져옴
#DB의 member table에 있는 내용을 읽어들여서,
# member.csv 파일에 저장해주세요
# aaa, 111 : join
'''
aaa, 111
bbb, 222
ccc, 333
'''

import pymysql
# DB 연결
outFp = open("member.csv", "w", encoding="utf-8")
conn = pymysql.connect(host="localhost", db="sqldb", user="root", password="1234")

# SQL 쿼리 실행
sql = "select * from member"
cur=conn.cursor()
cur.execute(sql)

# 결과를 가져와서 csv 파일에 저장
while True :
    line = cur.fetchone()   #1회성 할 일 
    if line == None :
        break
    outFp.writelines("f{line[0]}, {line[1]}" + "\n")

# 연결 닫기
outFp.close()
cur.close()
conn.close()
```
